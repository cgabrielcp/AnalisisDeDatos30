# Listas

```python
my_list_1 = [3, 5, 2]

print(my_list_1)
print(type(my_list_1))

my_list_2 = []
print(my_list_2)
print(type(my_list_2))

# Lista mixta
my_list_3 = [3, "Hola", 4.45]
print(my_list_3)
print(type(my_list_3))

# Lista Anidada
my_list_4 = ['apple', [3, ['a', 2, 2.5], 8], 'x']
print(my_list_4)
print(type(my_list_4))
```

```python
my_list_4.insert(1,30)
print(my_list_4)
print(type(my_list_4))
```

```python
my_list = ['a', 'b,' 'c', 'd', 'e', 'f', 'g', 'h', 'i']

print(my_list[2:5])

print(my_list[5:])
```

# Diccionario

```python
my_dict = {}

print(my_dict)


my_dict = dict({
    1: 'apple',
    2: 'pera'
})
print(my_dict)
```

```python
my_dict = {
    'name': 'Jack',
    'age': 26,
    'color': 'red'
}

print(my_dict['name'])

print(my_dict.get('color', 0))
```
