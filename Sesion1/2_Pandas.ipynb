# Agregaciones

```python
data = {
    'Grupo': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Valor': [10, 20, 15, 25, 12, 18]
}

# Convertirlo a df
df = pd.DataFrame(data)

promedio_por_grupo = df.groupby('Grupo').mean()
print(promedio_por_grupo)

suma_por_grupo = df.groupby('Grupo').sum()
print(suma_por_grupo)
```

```python
# Leemos un archivo sin extension y con separador distinto a la coma
df = pd.read_csv('utils/data/data', sep= '|')

df
```

```python
# Eliminar filas que contiene Nan ( desconocidos )
df_sin_nan = df.dropna()
print(df_sin_nan)

print('\n')

# reemplazar los valores faltantes
df_completo = df.fillna(value = 0)
print(df_completo)

print('\n')

# Eliminar solo las filas donde columna 'ciudad' tenga valores faltantes
df_sin_na_ciudad = df.dropna(subset=['ciudad'])
print(df_sin_na_ciudad)

print('\n')
# Reemplara faltantes en una sola columna
df_completo_edad = df['edad'].fillna(value=0)
print(df_completo_edad)
```

# Funciones en un Dataframe

```python
data = {
    'A': [1, 2, 3, 4],
    'B': [6, 7, 8, 9]
}

df = pd.DataFrame(data)
```

```python
# Crear una columna que multiplique por 2 los valores de A

df['A_doble'] = df['A'].apply(lambda x: x*2)

df
```

```python
def multiplodos(x):
    return x*2

df['A_doble'] = df['A'].apply(multiplodos)

df
```

```python
# Suma de las columnas A y B
df['Suma_metodo1'] = df.apply(lambda row: row['A'] + row['B'], axis=1)

df
```

```python
def suma(row):
    return row['A'] + row['B']

# Suma de las columnas A y B
df['Suma_metodo2'] = df.apply(suma, axis=1)

df
```
